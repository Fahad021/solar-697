```python
import geopandas
import json
import numpy as np
import pandas as pd

from geojson import dump
```

```python
shp_file = geopandas.read_file("../data/cb_2019_us_zcta510_500k.shp")
shp_file.to_file('zipcodes.geojson', driver='GeoJSON')
```

```python
with open("zipcodes.geojson") as f:
    zips = json.load(f)
    
for i, feature in enumerate(zips["features"]):
    zips["features"][i]["id"] = feature["properties"]['ZCTA5CE10']
zips["features"][0]
```

```python
with open("../data/plotly_zips.geojson", 'w') as outfile:
    dump(zips, outfile)
```
