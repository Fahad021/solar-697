```python
import json
import numpy as np
import pandas as pd
import pickle



import sqlite3
```

```python
with open("cfg.json", "r") as jin:
    cfg = json.loads(jin.read())

# rewrite to ensure formatting
with open("cfg.json", "w") as jout:
    json.dump(cfg, jout, indent=4)
```

```python
with open(cfg["data_dir"] + cfg["p_dates"], 'rb') as fp:
    prog_map = pickle.load(fp)
```

```python

conn_proj = sqlite3.connect(cfg["db_dir"] + cfg["install_db"])
cursor_proj = conn_proj.cursor()

query_frame = """
SELECT *

FROM installation

"""


df = pd.read_sql_query(query_frame, conn_proj)


conn_proj.close
```

```python
keep_cols = ['size', 'price','discount','housing']
keep_cols_2 = ['p_type_31','p_type_32', 'p_type_33', 'p_type_37',
             'p_type_46', 'p_type_50', 'p_type_78', 'p_type_87', 'p_type_88', 'p_type_92', 'p_type_86']

g_2 = df.groupby(['month','zipcode'])[keep_cols].mean().reset_index()

df['size'] = df['size'].astype('bool')
grouped_df = df.groupby(['month','zipcode'])['size'].sum().reset_index().rename(
    columns = {'size':'i_count'})


grouped_df = grouped_df.merge(g_2, left_on=['zipcode','month'], right_on=['zipcode','month'])

grouped_df.head()
```

```python
prog_dur_map = {87:1,88:18}


for program in [31,32,33,37,46,50,78,87,88,92,86]:
    
    def in_effect(x):
        
        z = x[0]
        d = x[1]
        
        p_type = program

        output = 0

        try:
            items = []
            for item in prog_map[p_type][z]:
                items.append(item[1])
                if output != 1:
                    #programs 87 and 88 expected to expire, correlated with Berkeley rebate_or_grant column
                    if program == 87 or program == 88:
                        if item[1] <= d <= item[1] + prog_dur_map[program]:
                            output = 1
                        elif (item[1] - prog_dur_map[program]) <= d <= item[1]:
                            output = -1
                    #other items assumed to not expire
                    else:
                        if item[1] <= d:
                            output = 1
                    #elif d == item[1]:
                    #    output = .5
            if d < min(items):
                output = -1

        except:
            output = -1
        
        return output
    
    grouped_df['p_type_' + str(program)] = [in_effect(x) for x in zip(grouped_df.zipcode,grouped_df.month)]
    print(str(program) + ' complete')
```

```python
conn_proj = sqlite3.connect(cfg["db_dir"] + cfg["install_db"])
cursor_proj = conn_proj.cursor()

grouped_df.to_sql("grouped_policy", conn_proj, if_exists="replace", index=False)

conn_proj.commit
```
